```python
import calendar
import pandas as pd
import numpy as np
```

```python

anio_target = 2026
month_target = 2
matrix_target = calendar.monthcalendar(anio_target, month_target)

anios = [anio for anio in range(2024, 3000)]
similar_years = []

for anio in anios:
    current_matrix = calendar.monthcalendar(anio, month_target)    
    if current_matrix == matrix_target:
        similar_years.append(anio)
```

```python
print(similar_years)
```

```python
intervals = np.diff(similar_years)
mean = intervals.mean()
```
