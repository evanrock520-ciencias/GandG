```python
from graph import Graph, Vertex, Edge
from controller import Controller
```

```python
def social_network(persons : list):
    vertices = [Vertex(person, set()) for person in persons]
    social_graph = Graph(vertices, [])
    
    for idx, person in enumerate(persons):
        for jdx, other in enumerate(persons):
            if person == other: 
                continue
            
            if person[0] == other[0]:
                social_graph.add_connection(idx, jdx)
                
    return social_graph
```

```python
social_graph = social_network(["Fernanda", "Carlos", "Carla", "Filiberto", "Cameron"])
social_graph.show_edges()
```

```python
social_graph.show_degree_sequence()
```
