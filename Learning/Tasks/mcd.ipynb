```python
from graph import Graph, Vertex, Edge
from controller import Controller
from math import gcd
```

```python
def mcd_graph(vertices: list):
    vtx_list = [Vertex(value, set()) for value in vertices]
    graph = Graph(vtx_list, [])
    
    for idx, vtx in enumerate(vtx_list):
        for jdx, utx in enumerate(vtx_list):
            if vtx == utx: continue
            
            if gcd(vtx.value, utx.value) == 1:
                graph.add_connection(idx, jdx)
                
    return graph
```

```python
graph = mcd_graph([2,3,4,5,11,12,13,14,18])
graph.show_edges()
```

```python
delta_min = graph.get_gamma_degree()
delta_max = graph.get_delta_degree()

delta_min, delta_max
```

```python
vertices = [idx for idx in range(1, 11)]
vertices
```

```python
def graph_abs(vertices: list):
    vtx_list = [Vertex(idx, set()) for idx in vertices]
    graph = Graph(vtx_list, [])
    
    for idx, vtx in enumerate(vtx_list):
        for jdx, utx in enumerate(vtx_list):
            if vtx == utx: continue
            
            if abs(vtx.value - utx.value) <= 3:
                graph.add_connection(idx, jdx)
                
    return graph
```

```python
abs_graph = graph_abs(vertices)
abs_graph.show_edges()
```

```python
tdg = abs_graph.get_total_degree()
odds = len([number for number in abs_graph.degree_sequence if number % 2 != 0])

tdg, odds
```
