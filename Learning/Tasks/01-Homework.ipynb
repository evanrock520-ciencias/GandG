# Gráficas isomorfas


Definición:
Dadas dos gráficas $G = (V (G), E(G))$ y $H = (V (H), A(H))$

Diremos que $G$ es isomorfa a $H$ si existe una función biyectiva
$$\phi : V (G) \to V (H)$$

Tal que cumple que:
1. Si $\{u, v\} \in A(G) \implies {\phi(u), \phi(v)} \in A(H)$
2. Si $\{u, v\} \notin A(G) \implies {\phi(u), \phi(v)} \notin A(H)$


26) Demostrar que, salvo isomorfismo, para cualquier n ∈ N \ {0} la grafica completa de orden n es única.


Sean $G$ y $H$ dos gráficas tales que:

- $|V(G)| = |V(H)| = n$
- Ambas son completas

Dado que $|V(G)| = |V(H)|$, existe una biyección
$$
f: V(G) \to V(H).
$$

Como $G$ es completa,
$$
\forall u,v \in V(G),\; u \neq v \implies \{u,v\} \in E(G).
$$

Sea $u,v \in V(G)$ con $u \neq v$. Entonces $f(u) \neq f(v)$, y como $H$ es completa,
$$
\{f(u),f(v)\} \in E(H).
$$

Por lo tanto,
$$
\{u,v\} \in E(G) \iff \{f(u),f(v)\} \in E(H),
$$
y así $f$ es un isomorfismo.

$\therefore G \cong H$.

29. Demostrar que $L(K_{1,3}) \cong L(K_3)$

Antes de realizar la prueba, intentemos comprobarlo mediante código.

```python
def complete(degree: int) -> Graph:
    vertices = [Vertex(idx, set()) for idx in range(degree)]
    graph = Graph(vertices, [])
    
    for idx, vtx in enumerate(graph.vertices):
        for jdx, utx in enumerate(graph.vertices):
            if vtx == utx: continue
            
            graph.add_connection(idx, jdx)
            
    return graph
```

Notemos que:
- $|V(K_3)| = 3$
- $|A(K_3)| = 6$
- El vector de grados de $K_3$ es de $(2,2,2)$

```python
graphK3 = complete(3)
graphK3.show_degrees()
```

```python
def complete_bipartite(m: int, n: int) -> Graph:
    vertices_x = [Vertex(idx, set()) for idx in range(m)]
    vertices_y = [Vertex(idx + m, set()) for idx in range(n)]
    
    all_vertices = vertices_x + vertices_y
    graph = Graph(all_vertices, [])
    
    for idx, vtx in enumerate(vertices_x):
        for jdx, utx in enumerate(vertices_y):
            graph.add_connection(idx, m + jdx)
            
    return graph
```

Y en este caso:
- $|V(K_{1,3})| = 4$
- $|A(K_{1,3})| = 3$
- El vector de grados de $K_3$ es de $(3,1,1,1)$

```python
graphK13 = complete_bipartite(1, 3)
graphK13.show_degrees()
```

Veamos que si $L(K_{1,3}) \cong L(K_3)$ entonces se cumplen las siguientes propiedades como mínimo:
- $|V(L(K_{1,3}))| = |V(L(K_3))|$
- $|A(L(K_{1,3}))| = |A(L(K_3))|$
- La secuencia de grados de $L(K_{1,3})$ es igual a la secuencia de grados $L(K_3)$

Dado que sus tamaños son iguales $\implies$
$$\exists \; \phi : V(L(K_3)) \to V(L(K_{1,3}))$$

Construyendola:

- $\phi (a_1) = e_1$ 
- $\phi (a_2) = e_2$ 
- $\phi (a_3) = e_3$ 

Que ademas preserva las adyacencias que existian. Pues:

- $\{a_1, a_2\} \in V(L(K_3))$ 
- $\{a_2, a_3\} \in V(L(K_3))$

- $\{\phi (a_1) = e_1, \phi(a_2) = e_2\} \in V(L(K_{1_3}))$

$\therefore L(K_3) \cong L(K_{1,3})$
